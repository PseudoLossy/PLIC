```python
import os
from image_similarity_measures.evaluate import evaluation
import cv2 as cv
import numpy as np
```

```python
color_input = 'color_dataset'
color_output = 'color_recon'

color_dataset = [file for file in os.listdir(color_input)]
color_dataset = ["./" + color_input + "/" + x for x in color_dataset]

color_recon = [file for file in os.listdir(color_output)]
color_recon = ["./" + color_output + "/" + x for x in color_recon]
```

```python
ssim_scores = []

for i in range(149):
    print(i)

    results = evaluation(color_dataset[i], color_recon[i], metrics=["ssim"])

    ssim_scores.append(results['ssim'])
```

```python
print("USING SSIM TO COMPARE COLOR MAPS:")
print("Mean: ", np.mean(ssim_scores))
print("Max: ", np.max(ssim_scores))
print("Min: ", np.min(ssim_scores))
print("Std: ", np.std(ssim_scores))
print("Median: ", np.median(ssim_scores))
```
